### Model training

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import f1_score, balanced_accuracy_score
import matplotlib.pyplot as plt
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# Download resources
nltk.download('punkt')
nltk.download('stopwords')

def preprocess_text(text):
    """
    Preprocesses text data for aphasia detection model.

    Steps:
    - Lowercasing
    - Removing punctuation (except for repetitions)
    - Tokenizing text
    - Flagging word repetitions

    :param text: Raw input text (string)
    :return: Cleaned text with repetitions flagged
    """
    # Lowercase the text
    text = text.lower()

    # Remove punctuation (if needed, retain repetition indicators)
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Tokenize text into words
    words = word_tokenize(text)

    # Remove stopwords (optional)
    #stop_words = set(stopwords.words('english'))
    #words = [word for word in words if word not in stop_words]

    # Flag repeated words (for analysis of aphasic patterns)
    repetitions = flag_repetitions(words)

    #Return a string or tokenized words with repetition counts
    return ' '.join(repetitions)

def flag_repetitions(words):
    """
    Flags consecutive repeated words, adding markers like '_repeated'.

    :param words: List of tokenized words
    :return: List of words with flagged repetitions
    """
    flagged_words = []
    previous_word = None
    repetition_count = 0

    for word in words:
        if word == previous_word:
            repetition_count += 1
        else:
            repetition_count = 0

        if repetition_count > 0:
            flagged_words.append(f"{word}_repeated")  # Flag the repetition
        else:
            flagged_words.append(word)

        previous_word = word

    return flagged_words

# Example
raw_text = "I I I went to to the the store and and bought bread"
processed_text = preprocess_text(raw_text)
print(processed_text)

```

OneVsAll training using TF-IDF Vectorisation and standard evaluation.

```python
X = df_expanded["Text"]
y = df_expanded["Subtype"]

# Initialize LabelEncoder and transform the labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Print the classes detected by LabelEncoder
print("Classes found by LabelEncoder:")
print(le.classes_)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Preprocess the text data
X_train_preprocessed = X_train.apply(preprocess_text)
X_test_preprocessed = X_test.apply(preprocess_text)

# Feature extraction with TF-IDF
vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train_preprocessed)
X_test_tfidf = vectorizer.transform(X_test_preprocessed)

# Classifier pool
classifiers = {
    "MultinomialNB": MultinomialNB(),
    "DecisionTree": DecisionTreeClassifier(),
    "RandomForest": RandomForestClassifier(),
    "GradientBoosting": GradientBoostingClassifier(),
    "LogisticRegression": LogisticRegression(),
    "SVM": SVC(C=10, gamma='scale', kernel='linear', probability=True),
    "MLPClassifier": MLPClassifier(max_iter=1000)
}

# Initialize dataframes for storing scores
f1_scores_df = pd.DataFrame(index=le.classes_, columns=classifiers.keys())
balanced_accuracy_scores_df = pd.DataFrame(index=le.classes_, columns=classifiers.keys())

# Iterate over each label to train classifiers and evaluate performance
for label in le.classes_:
    y_train_label = (y_train == le.transform([label])[0]).astype(int)
    y_test_label = (y_test == le.transform([label])[0]).astype(int)

    for name, clf in classifiers.items():
        clf.fit(X_train_tfidf, y_train_label)
        y_pred = clf.predict(X_test_tfidf)

        # Calculate and store the F1-score and balanced accuracy
        f1_score_val = f1_score(y_test_label, y_pred, average='binary')
        balanced_accuracy_val = balanced_accuracy_score(y_test_label, y_pred)
        f1_scores_df.loc[label, name] = f1_score_val
        balanced_accuracy_scores_df.loc[label, name] = balanced_accuracy_val

        # Print the scores for the current classifier and label
        print(f"F1-score for '{label}' using '{name}': {f1_score_val:.2f}, Balanced Accuracy: {balanced_accuracy_val:.2f}")

# Optionally, display the summary of best scores and classifiers
print("\nF1-scores DataFrame:\n", f1_scores_df)
print("\nBalanced Accuracy Scores DataFrame:\n", balanced_accuracy_scores_df)
```

Same model as above, evaluation using a 5 fold cross validation (stratified)

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score, balanced_accuracy_score

# Initialize StratifiedKFold with 5 splits
skf = StratifiedKFold(n_splits=5)

# Initialize dataframes to store the average scores across folds
f1_scores_df = pd.DataFrame(index=le.classes_, columns=classifiers.keys())
balanced_accuracy_scores_df = pd.DataFrame(index=le.classes_, columns=classifiers.keys())

# Initialize dict to keep track of cumulative scores
cumulative_f1_scores = {name: np.zeros(len(le.classes_)) for name in classifiers.keys()}
cumulative_balanced_accuracy_scores = {name: np.zeros(len(le.classes_)) for name in classifiers.keys()}

# Iterate over each fold
for fold, (train_idx, test_idx) in enumerate(skf.split(X, y_encoded)):
    print(f"Fold {fold+1}")

    # Split data into train and test sets for this fold
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y_encoded[train_idx], y_encoded[test_idx]

    # Preprocess the text data
    X_train_preprocessed = X_train.apply(preprocess_text)
    X_test_preprocessed = X_test.apply(preprocess_text)

    # Feature extraction with TF-IDF
    vectorizer = TfidfVectorizer(max_features=5000)
    X_train_tfidf = vectorizer.fit_transform(X_train_preprocessed)
    X_test_tfidf = vectorizer.transform(X_test_preprocessed)

    # Iterate over each label to train classifiers and evaluate performance
    for label in le.classes_:
        y_train_label = (y_train == le.transform([label])[0]).astype(int)
        y_test_label = (y_test == le.transform([label])[0]).astype(int)

        for name, clf in classifiers.items():
            clf.fit(X_train_tfidf, y_train_label)
            y_pred = clf.predict(X_test_tfidf)

            # Calculate F1-score and balanced accuracy
            f1_score_val = f1_score(y_test_label, y_pred, average='binary')
            balanced_accuracy_val = balanced_accuracy_score(y_test_label, y_pred)

            # Accumulate the scores across folds
            cumulative_f1_scores[name][le.transform([label])[0]] += f1_score_val
            cumulative_balanced_accuracy_scores[name][le.transform([label])[0]] += balanced_accuracy_val

# Average the scores across the 5 folds
for name in classifiers.keys():
    f1_scores_df[name] = cumulative_f1_scores[name] / skf.get_n_splits()
    balanced_accuracy_scores_df[name] = cumulative_balanced_accuracy_scores[name] / skf.get_n_splits()

# Display the final averaged scores
print("\nF1-scores DataFrame:\n", f1_scores_df)
print("\nBalanced Accuracy Scores DataFrame:\n", balanced_accuracy_scores_df)
```

Same model training evaluated using a normal 5 fold cross validation

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report
import pandas as pd

# Assuming df_expanded is your DataFrame
X = df_expanded["Text"]
y = df_expanded["Subtype"]

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Print the classes detected by LabelEncoder
print("Classes found by LabelEncoder:")
print(le.classes_)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Preprocess the text data
X_train_preprocessed = X_train.apply(preprocess_text)
X_test_preprocessed = X_test.apply(preprocess_text)

# Feature extraction with TF-IDF
vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train_preprocessed)
X_test_tfidf = vectorizer.transform(X_test_preprocessed)

# Classifier pool
classifiers = {
    "MultinomialNB": MultinomialNB(),
    "DecisionTree": DecisionTreeClassifier(),
    "RandomForest": RandomForestClassifier(),
    "GradientBoosting": GradientBoostingClassifier(),
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "SVM": SVC(C=10, gamma='scale', kernel='linear', probability=True),
    "MLPClassifier": MLPClassifier(max_iter=1000)
}

# Initialize DataFrame for storing scores
evaluation_metrics = pd.DataFrame(columns=["Classifier", "Train Accuracy", "Test Accuracy", "Macro F1-score (Test)", "CV Accuracy"])

# Iterate over classifiers
for name, clf in classifiers.items():
    # Train the classifier on the training set
    clf.fit(X_train_tfidf, y_train)

    # Predict on the test set
    y_pred_test = clf.predict(X_test_tfidf)

    # Calculate evaluation metrics on the test set
    test_accuracy = accuracy_score(y_test, y_pred_test)
    test_macro_f1 = f1_score(y_test, y_pred_test, average='macro')

    # Perform 5-fold cross-validation on the training set
    cv_accuracy = cross_val_score(clf, vectorizer.transform(X.apply(preprocess_text)), y_encoded, cv=5, scoring='accuracy')
    avg_cv_accuracy = cv_accuracy.mean()

    # Append metrics to DataFrame
    evaluation_metrics = pd.concat(
        [evaluation_metrics, pd.DataFrame({
            "Classifier": [name],
            "Train Accuracy": [clf.score(X_train_tfidf, y_train)],
            "Test Accuracy": [test_accuracy],
            "Macro F1-score (Test)": [test_macro_f1],
            "CV Accuracy": [avg_cv_accuracy]
        })],
        ignore_index=True
    )

    # Print classification report
    print(f"\nClassifier: {name}")
    print(classification_report(y_test, y_pred_test, target_names=le.classes_))

# Display summary of evaluation metrics
print("\nEvaluation Metrics:\n", evaluation_metrics)
```

Same model evaluated using a 5 fold cross validation but with word2vec embeddings.

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report
from gensim.models import Word2Vec
import pandas as pd
import numpy as np

# Assuming df_expanded is your DataFrame
X = df_expanded["Text"]
y = df_expanded["Subtype"]

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Print the classes detected by LabelEncoder
print("Classes found by LabelEncoder:")
print(le.classes_)

# Preprocess the text data
X_preprocessed = X.apply(preprocess_text)

# Tokenize the preprocessed text
X_tokenized = [text.split() for text in X_preprocessed]

# Train Word2Vec model
word2vec_model = Word2Vec(sentences=X_tokenized, vector_size=100, window=5, min_count=1, workers=4)

# Function to average word vectors for each document
def document_vector(doc):
    # Remove out-of-vocabulary words
    doc = [word for word in doc if word in word2vec_model.wv.key_to_index]
    return np.mean(word2vec_model.wv[doc], axis=0) if doc else np.zeros(word2vec_model.vector_size)

# Create document vectors
X_vectors = np.array([document_vector(doc) for doc in X_tokenized])

# Classifier pool
classifiers = {
    "DecisionTree": DecisionTreeClassifier(),
    "RandomForest": RandomForestClassifier(),
    "GradientBoosting": GradientBoostingClassifier(),
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "SVM": SVC(C=10, gamma='scale', kernel='linear', probability=True),
    "MLPClassifier": MLPClassifier(max_iter=1000)
}

# Initialize DataFrame for storing scores
evaluation_metrics = pd.DataFrame(columns=["Classifier", "Accuracy", "Macro F1-score"])

# Iterate over classifiers
for name, clf in classifiers.items():
    # Perform 5-fold cross-validation
    cv_accuracy = cross_val_score(clf, X_vectors, y_encoded, cv=5, scoring='accuracy')
    cv_macro_f1 = cross_val_score(clf, X_vectors, y_encoded, cv=5, scoring='f1_macro')

    # Calculate average scores
    avg_accuracy = cv_accuracy.mean()
    avg_macro_f1 = cv_macro_f1.mean()

    # Append metrics to DataFrame
    evaluation_metrics = pd.concat([evaluation_metrics, pd.DataFrame({"Classifier":[name], "Accuracy":[avg_accuracy], "Macro F1-score":[avg_macro_f1]})], ignore_index=True)

    # Print classifier metrics
    print(f"\nClassifier: {name}")
    print(f"Average Accuracy: {avg_accuracy:.4f}")
    print(f"Average Macro F1-score: {avg_macro_f1:.4f}")

# Display summary of evaluation metrics
print("\nEvaluation Metrics:\n", evaluation_metrics)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the Test Accuracy, Macro F1-score, and CV Accuracy
plt.figure(figsize=(12, 8))

# Test Accuracy
plt.subplot(2, 1, 1)
sns.barplot(x="Classifier", y="Accuracy", data=evaluation_metrics, palette="Blues_d") # 'Test Accuracy' column exists in evaluation_metrics DataFrame
plt.title("Test Accuracy Comparison")
plt.xticks(rotation=45)

# Macro F1-score
plt.subplot(2, 1, 2)
sns.barplot(x="Classifier", y="Macro F1-score", data=evaluation_metrics, palette="Greens_d")
plt.title("Macro F1-score Comparison")
plt.xticks(rotation=45)

# Adjust the layout to prevent overlap
plt.tight_layout()
plt.show()
```
